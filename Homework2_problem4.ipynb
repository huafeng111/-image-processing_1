```python
import matplotlib.pyplot as plt
import copy
import numpy as np
import copy
from sklearn.metrics import mean_squared_error
```

```python
img1 = plt.imread('geisel.jpg') # Read an image
plt.imshow(img1) # Show the original image
plt.show()
```

```python
img1=np.pad(img1,((1,1),(1,1),(0,0)),"symmetric")
```

```python
def quantize(img):
    im=copy.deepcopy(img)
    for y in range(1,im.shape[0]-1):
        for x in range(1,im.shape[1]-1):
            old_pixel=copy.deepcopy(im[y][x])
            new_pixel=find_closest_color(old_pixel)
            im[y][x]=new_pixel
    return im
```

```python
def find_closest_color(pixel):
    new_pixel=copy.deepcopy(pixel)
    for i in range(3):
        new_pixel[i]=np.floor(pixel[i]*10/255)*(255/10)
    return new_pixel
    
```

```python
def reverse(pix):
    return np.where(pix>255,255,pix)
```

```python
def fsd(im2):
    im=copy.deepcopy(im2)
    for y in range(1,im.shape[0]-1):
        for x in range(1,im.shape[1]-1):
            old_pixel=copy.deepcopy(im[y][x])
            new_pixel=find_closest_color(old_pixel)
            im[y][x]=new_pixel
            quant_error=old_pixel- new_pixel
            im[y][x+1]= reverse(im[y][x+1]+quant_error*(7/16))
            im[y+1][x-1]= reverse( im[y+1][x-1]+quant_error*(3/16))
            im[y+1][x]= reverse(im[y+1][x]+quant_error*(5/16))
    return im
            
    
```

```python
img1=np.pad(img1,((1,1),(1,1),(0,0)),"symmetric")
new_img=fsd(img1)
new_img2=quantize(img1)
plt.imshow(new_img) # Show the original image
plt.show()
plt.imshow(new_img) # Show the ditering image
plt.show()
plt.imshow(new_img2) # Show the quantize image
plt.show()
```
