```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import copy
```

```python
def ahe(new_img,winsize):
    for i in range(new_img.shape[0]-winsize):
        for j in range(new_img.shape[1]-winsize):
            rank=0
            start_row=i+(winsize-1)//2
            start_col=j+(winsize-1)//2
            window=r_window(start_row,start_col,new_img,winsize)
            if new_img[start_row][start_col]==1.0:
                new_img[start_row][start_col]=255
            else:
                for z in range(winsize):
                    for g in range(winsize):
                            if new_img[start_row][start_col]>window[z][g]:
                                    rank+=1
                new_img[start_row][start_col]=rank*255/(winsize*winsize)
    plt.imshow(new_img,cmap="gray") # Show the original image
    return new_img
```

```python
def r_window(row,col,img,winsize):
    window=[]
    for i in range(winsize):
        a=img[row-(winsize-1)//2+i][col-(winsize-1)//2:col+(winsize-1)//2+1]
        window.append(a)
    window=np.array(window)
    return window
    
```

```python
img = plt.imread('beach.png') # Read an image
plt.imshow(img,cmap="gray") # Show the original image
plt.show()
```

```python
img_129 =np.pad(img ,((64,64),(64,64)),'reflect')
img_65 =np.pad(img ,((32,32),(32,32)),'reflect')
img_33 =np.pad(img ,((16,16),(16,16)),'reflect')
```

```python
img33=ahe(img_33,33)
plt.imshow(img33,cmap="gray") 
plt.show()
```

```python
img65=ahe(img_65,65)
plt.imshow(img65,cmap="gray") # Show the original image
plt.show()
```
